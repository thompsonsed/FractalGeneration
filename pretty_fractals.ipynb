# Generate a few pretty images at medium sizes

```python
import os
from fractal_generator import FractalGenerator
```

```python
# Define the folder for outputting fractal images
output_dir = "images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
start_y = -0.087
end_y = -0.086
start_x = 0.6545
end_x = 0.6565
res_x = 2000
res_y = 1000
r1, r2, r3 = 0.5, 0.6, 0.6
m = FractalGenerator()
m.set_grid(start_x, end_x, 
           start_y, end_y, 
           res_x, res_y, threshold=4)
m.generate_mandelbrot(iterations=200)
im = m.get_coloured_grid(r1, r2, r3)
im.save(os.path.join(output_dir, "mandelbrot_grey_1.png"))
```

```python
start_x = -0.40
end_x = -0.29
start_y = 0.345
end_y = 0.5
res_x = 800
res_y = 1200
r1, r2, r3 = 0.5, 0.1, 0.5
m = FractalGenerator()
m.set_grid(start_x, end_x, 
           start_y, end_y, 
           res_x, res_y, threshold=3)
m.generate_mandelbrot(iterations=150)
im = m.get_coloured_grid(r1, r2, r3, b1=3, b2=0.2, b3=3)
im = im.rotate(90, expand=1)
im.save(os.path.join(output_dir, "mandelbrot_green_1.png"))
```

```python
start_y = -0.746-0.0016875 
end_y = -0.746
start_x = 0.099
end_x = 0.102
res_x = 1920
res_y = 1080
threshold = 5
# Define the colour ratios
r1, r2, r3 = 0.3, 0.8, 0.1
m = FractalGenerator()
m.set_grid(start_x, end_x, start_y, end_y, res_x, res_y, threshold)
m.generate_mandelbrot(iterations=300)
im = m.get_coloured_grid(r1, r2, r3)
im.save(os.path.join(output_dir, "mandelbrot_purple.png"))
```
